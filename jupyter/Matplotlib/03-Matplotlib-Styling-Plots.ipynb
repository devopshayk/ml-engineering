### Данные

```python
import numpy as np
```

```python
x = np.arange(0,10)
y = 2 * x
```

## Легенда

С помощью именованного параметра **label="label text"** можно указать названия для тех или иных линий на графике. Далее с помощью метода **legend**, уже без параметров, можно добавить легенду: 

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x, x**2, label="x**2")
ax.plot(x, x**3, label="x**3")
ax.legend()
```

Обратите внимание, что легенда может потенциально загородить собой часть линий!

В методе **legend** есть опциональный именованный параметр **loc**, с помощью которого можно указать, где именно рисовать легенду. Возможные значения параметра **loc** - это числа, каждое из которых соответствует определённому месту на графике. Более подробно можно почитать [здесь](http://matplotlib.org/users/legend_guide.html#legend-location). Наиболее частые значения **loc** следующие:

```python
# Много различных вариантов....

ax.legend(loc=1) # верхний правый угол
ax.legend(loc=2) # верхний левый угол
ax.legend(loc=3) # нижний левый угол
ax.legend(loc=4) # нижний правый угол

# .. доступны также другие варианты

# Хороший вариант
ax.legend(loc=0) # matplotlib сам выбирает оптимальное местоположение
fig
```

```python
ax.legend(loc=(1.1,0.5)) # задаём вручную
fig
```

## Цвета, ширина линий и типы линий

Matplotlib предоставляет *много* возможностей по настройке цветов, ширины линий и типов линий 

Для этого можно использовать базовый синтаксис, похожий на MATLAB (именно MATLAB, а не Matplotlib). Но этот синтаксис не очень удобен, и лучше избегать его. Ну разве что если Вы раньше работали с MATLAB, тогда да. Вместо этого давайте сосредоточимся на именованных параметрах.

### Быстрый обзор:

### Цвета с использованием синтаксиса MatLab

В matplotlib можно задавать цвета линий и других графических элементов разными способами. Первый способ - похожий на MATLAB синтаксис, в котором `'b'` означает blue, `'g'` означает green и так далее. Также поддерживается MATLAB API для выбора стиля линий: например, 'b.-' означает цвет blue и линия с точками:

```python
# Цвета и стили с применением синтаксиса MATLAB 
fig, ax = plt.subplots()
ax.plot(x, x**2, 'b.-') # blue line with dots
ax.plot(x, x**3, 'g--') # green dashed line
```

## Рекомендуемый способ: именованные параметры

### Для цветов есть параметр color

Второй способ - указывать цвета с помощью их названий, а также с помощью шестнадцатиричных кодов RGB. Опционально можно ещё указать прозрачность. Параметры называются `color` и `alpha`.

```python
fig, ax = plt.subplots()

ax.plot(x, x+1, color="blue", alpha=0.5) # полупрозрачный
ax.plot(x, x+2, color="#4000ff")        # RGB hex code
ax.plot(x, x+3, color="#FF8C00")        # RGB hex code 
```

## Стилизация линий и маркеров

## Ширина линий

Для изменения ширины линии есть именованный параметр `linewidth` или `lw`. 

```python
fig, ax = plt.subplots(figsize=(12,6))

# Пишем linewidth или lw
ax.plot(x, x-1, color="red", linewidth=0.25)
ax.plot(x, x-2, color="red", lw=0.50)
ax.plot(x, x-3, color="red", lw=3)
ax.plot(x, x-4, color="red", lw=10)
```

## Стили линий

Различных вариантов много, вот примеры:

```python
# Возможные варианты для linestype:  ‘--‘, ‘–’, ‘-.’, ‘:’, ‘steps’
fig, ax = plt.subplots(figsize=(12,6))

ax.plot(x, x-1, color="green", lw=3, linestyle='-') # solid
ax.plot(x, x-2, color="green", lw=3, ls='-.') # dash and dot
ax.plot(x, x-3, color="green", lw=3, ls=':') # dots
ax.plot(x, x-4, color="green", lw=3, ls='--') # dashes
```

### Произвольная стилизация линии

При желании можно задать любую последовательность отрезков линии и пробелов между ними. Для этого можно указать длину каждого отрезка (в количестве точек) и длину пробела между каждым из отрезков (тоже в количестве точек).

Например, (5, 2, 1, 2) будет означать следующее:

    отрезок линии длиной в 5 точек
    затем пробел длиной в 2 точек
    далее отрезок линии в 1 точку
    и наконец, пробел длиной в 2 точки.

Для начала посмотрим, как мы можем "получить" линию с помощью команды .plot():

```python
fig, ax = plt.subplots(figsize=(12,6))

lines = ax.plot(x,x)

print(type(lines))
```

```python
fig, ax = plt.subplots(figsize=(12,6))
# произвольная стилизация линии
lines = ax.plot(x, x+8, color="black", lw=5)
lines[0].set_dashes([10, 10]) # формат: длина линии, длина пробела. 
```

```python
fig, ax = plt.subplots(figsize=(12,6))
# произвольная стилизация линии
lines = ax.plot(x, x+8, color="black", lw=5)
lines[0].set_dashes([1,1, 1,1, 10,10]) # формат: длина линии, длина пробела и т.д.
```

## Маркеры

По сути мы раньше всегда рисовали точки, и matplotlib автоматически соединял эти точки с помощью линий. Давайте посмотрим, как мы можем нарисовать маркеры в каждой из этих точек.

### Стили маркеров

Большой список различных типов маркеров можно найти здесь: https://matplotlib.org/3.2.2/api/markers_api.html

```python
fig, ax = plt.subplots(figsize=(12,6))

# Параметр marker содержит строковый код
# Параметр markersize или ms задаёт размер


ax.plot(x, x-1,marker='+',markersize=20)
ax.plot(x, x-2,marker='o',ms=20) #ms = markersize
ax.plot(x, x-3,marker='s',ms=20,lw=0) # указываем linewidth=0, чтобы увидеть только маркеры
ax.plot(x, x-4,marker='1',ms=20)
```

### Различные края маркеров, ширина, размер и стиль

```python
fig, ax = plt.subplots(figsize=(12,6))

# размер и цвет маркеров
ax.plot(x, x, color="black", lw=1, ls='-', marker='s', markersize=20, 
        markerfacecolor="red", markeredgewidth=8, markeredgecolor="blue");
```
